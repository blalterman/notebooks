```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Define dispersion relation and constants

```python
E0 = 10
t1 = 2
t2 = 4
a1 = 1
a2 = 0.5
def dispersion(kx, ky):
    return E0 - (2 * t1 * np.cos(kx * a1)) - (2 * t2 * np.cos(ky * a2))
```

- Create an array of $k_{x}$ and $k_{y}$ points
- Evaluate the dispersion relation at each point

```python
npoints = 1000
kxs = np.linspace(-2 * np.pi / a1, 2 * np.pi / a1, npoints)
kys = np.linspace(2 * -np.pi / a2, 2 * np.pi / a2, npoints)
kxs, kys = np.meshgrid(kxs, kys)
E = dispersion(kxs, kys)
```

Draw a contour plot of electron energies as a function of $k_{x}$ and $k_{y}$
- Darker blues are lower energies
- Lighter yellows are higher energies
- The 1st Brillouin Zone is the rectangle in the middle

Now it's clear to see that the lowest energy states are in the centre of the Brilliouin zone, and the highest energies are at the corners.

```python
fig, ax = plt.subplots(figsize=(12,8))
cplot = ax.contourf(kxs, kys, E, 20)
fig.colorbar(cplot, label='Energy')

ax.set_xlabel('$k_{x}$', fontsize=16)
ax.set_ylabel('$k_{y}$', fontsize=16)
ax.axvline(np.pi / a1, color='r')
ax.axvline(-np.pi / a1, color='r')
ax.axhline(np.pi / a2, color='r')
ax.axhline(-np.pi / a2, color='r')
```
