```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
from spectrum import rgb
import numpy as np
```

```python
# Outgoing angle as a function of ingoing angle
def theta(alpha, eta):
    return 4 * np.arcsin(np.sin(alpha) / eta) - 2 * alpha
```

```python
# Plot theta for a range of alpha
alpha = np.linspace(0, np.pi / 2, 100)
eta = 1.33

fig, ax = plt.subplots()
ax.plot(np.rad2deg(alpha), np.rad2deg(theta(alpha, eta)))
ax.set_xlabel(r'$\alpha$' + ' (deg)')
ax.set_ylabel(r'$\theta$' + ' (deg)');
```

```python
# Define a range of wavelengths and corresponding refractive indices
# (data given in question)
wls = [650, 590, 570, 510, 475, 445, 400]
etas = np.array([1.330, 1.334, 1.335, 1.338, 1.341, 1.3425, 1.348])
colours = rgb(np.array(wls))

fig, ax = plt.subplots()
for (eta, colour) in zip(etas, colours):
    ax.plot(np.rad2deg(alpha), np.rad2deg(theta(alpha, eta)), c=colour)
ax.set_ylim(30, 45)
ax.set_xlim(40, 80)
ax.set_xlabel(r'$\alpha$' + ' (deg)')
ax.set_ylabel(r'$\theta$' + ' (deg)');
```

```python
# Angle at which rays cluster (ie. the theta value at which plots above peak)
def clusterangle(eta):
    alpha = np.arccos(np.sqrt((eta**2 - 1) / 3))
    return theta(alpha, eta)
```

```python
fig, ax = plt.subplots()
ax.scatter(wls, np.rad2deg(clusterangle(etas)), c=colours)
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Outgoing cluster angle (deg)');
```
